```python
import pandas as pd
```

```python
#read in Glassdoor data analyst csv from Kaggle
data_analyst = "Resources\DataAnalyst.csv"
data_analyst_df = pd.read_csv(data_analyst)
```

```python
data_analyst_df
```

```python
#filter and pick columns for final Glassdoor data analyst df
data_analyst_df.head()
data_analyst_transformed=data_analyst_df[["Job Title","Salary Estimate","Job Description","Company Name","Location"]]
data_analyst_transformed
```

```python
#add source column to data analyst df
data_analyst_transformed["Source"]="www.glassdoor.com"
data_analyst_transformed
```

```python
#remove (glassdoor est.) from salary estimate column
data_analyst_transformed["Salary Estimate"].str.replace(r"\(.*\)"," ")
```

```python
data_analyst_transformed["Salary Estimate"]=data_analyst_transformed["Salary Estimate"].str.replace(r"\(.*\)"," ")
data_analyst_transformed
```

```python
#read in dice.com tech jobs csv from Kaggle
dice_jobs = "Resources\dice_com-job_us_sample.csv"
dice_jobs_df = pd.read_csv(dice_jobs)
```

```python
#extract positions that included data analyst as part of job title
dice_df=dice_jobs_df[dice_jobs_df['jobtitle'].str.contains("Data Analyst")]
```

```python
#filter and pick columns for final dice.com df
#renamed columns to match data analyst df
dice_df_transformed= dice_df[["company", "jobdescription","joblocation_address","jobtitle"]]
dice_df_transformed=dice_df_transformed.rename(columns={"company":"Company Name","jobdescription":"Job Description","joblocation_address": "Location","jobtitle":"Job Title"})
```

```python
#add source and salary estimate to dice.com df
dice_df_transformed["Source"]="www.dice.com"
dice_df_transformed["Salary Estimate"]=""
dice_df_transformed
```

```python
# Remove \n* from company name
data_analyst_transformed["Company Name"]=data_analyst_transformed["Company Name"].str.split("\\n", expand=True)[0]
# data_analyst_transformed["Job Description"]=data_analyst_transformed["Job Description"].str.split("\\n", expand=True)[1]
data_analyst_transformed
```

```python
combined_data_analyst=dice_df_transformed.append(data_analyst_transformed)
combined_data_analyst_df = combined_data_analyst.reset_index(drop=True)
combined_data_analyst_df
```
