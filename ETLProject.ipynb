### Import & Transform Dice Data

```python
#read in dice.com tech jobs csv from Kaggle
dice_jobs = "Resources\dice_com-job_us_sample.csv"
dice_jobs_df = pd.read_csv(dice_jobs)
```

```python
#extract positions that included data analyst as part of job title
dice_df=dice_jobs_df[dice_jobs_df['jobtitle'].str.contains("Data Analyst")]
```

```python
#filter and pick columns for final dice.com df
#renamed columns to match data analyst df
dice_df_transformed= dice_df[["company", "jobdescription","joblocation_address","jobtitle"]]
dice_df_transformed=dice_df_transformed.rename(columns={"company":"Company Name",
                                                        "jobdescription":"Job Description",
                                                        "joblocation_address": "Location",
                                                        "jobtitle":"Job Title"})
```

```python
#add source and salary estimate to dice.com df
dice_df_transformed["Source"]="www.dice.com"
dice_df_transformed["Salary Estimate"]=""
dice_df_transformed.head()
```

```python
# Remove \n* from company name
data_analyst_transformed["Company Name"]=data_analyst_transformed["Company Name"].str.split("\\n", expand=True)[0]
# data_analyst_transformed["Job Description"]=data_analyst_transformed["Job Description"].str.split("\\n", expand=True)[1]
data_analyst_transformed.head()
```

### Combine the 2 sets of data

```python
combined_data_analyst=dice_df_transformed.append(data_analyst_transformed)
combined_data_analyst_df = combined_data_analyst.reset_index(drop=True)
combined_data_analyst_df.head()
```

### Create the normalized tables to be loaded into the database

```python
# Create tables to normalize the data to then load into a database
# Source

source_df = combined_data_analyst_df.groupby(['Source']).count()
source_df = source_df.reset_index()
source_df['id'] = source_df.index + 1
source_df = source_df[['id','Source']]
source_df = source_df.rename(columns={'Source':'source'})
source_df
```

```python
# Salary
salary_df = combined_data_analyst_df.groupby(['Salary Estimate']).count()
salary_df = salary_df.reset_index()
salary_df['id'] = salary_df.index + 1

salary_df = salary_df[['id','Salary Estimate']]
salary_df = salary_df.rename(columns={'Salary Estimate':'salary_estimate'})
salary_df.head()
```

```python
# Location
location_df = combined_data_analyst_df.groupby(['Location']).count()
location_df = location_df.reset_index()
location_df['id'] = location_df.index + 1

location_df = location_df[['id','Location']]
location_df = location_df.rename(columns={'Location':'location'})
location_df.head()
```

```python
# Bring in foreign keys
job_df = pd.merge(combined_data_analyst_df,source_df,left_on='Source', right_on='source',how='inner')
job_df = job_df.rename(columns={'id':'source_id',
                                'Company Name':'company_name',
                                'Job Description':'job_description',
                                'Job Title':'job_title'})

job_df = pd.merge(job_df, salary_df, left_on='Salary Estimate',right_on='salary_estimate',how='inner')
job_df = job_df.rename(columns={'id':'salary_id'})

job_df = pd.merge(job_df, location_df, left_on='Location', right_on='location',how='inner')
job_df = job_df.rename(columns={'id':'location_id'})

job_df = job_df[['company_name','job_title','job_description','source_id','salary_id','location_id']]
job_df.head()
```

### Connect and load data to database

```python
# Connect to local database
rds_connection_string = "usr1:password@localhost:5432/etlproject_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
# Add source to database
source_df.to_sql(name='source', con=engine, if_exists='append', index=False)
```

```python
# Add location to database
location_df.to_sql(name='location', con=engine, if_exists='append', index=False)
```

```python
# Add salary to database
salary_df.to_sql(name='salary', con=engine, if_exists='append', index=False)
```

```python
job_df.to_sql(name='job', con=engine, if_exists='append', index=False)
```
